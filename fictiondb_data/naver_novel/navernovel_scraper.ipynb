```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
from urllib.error import HTTPError
from urllib.error import URLError
from time import sleep


```

```python
class scraper:
    # options = webdriver.ChromeOptions()
    # options.add_experimental_option("excludeSwitches", ["enable-logging"])
    # driver = webdriver.Chrome(options=options, executable_path='C:/chromedriver_win32/chromedriver.exe')
    def naver(self):
        #완결부터
        self.naver_finish_novel()
        self.result = pd.DataFrame(zip(self.photourl, self.title,self.status,self.writer,self.status_num, self.pagelink))
    def naver_finish_novel(self):
        self.photourl = []
        self.title = []
        self.status = []
        self.writer = []
        self.status_num = []
        self.pagelink = []
        page_num = 1
        while True:
            from time import sleep
            allpage = []            
            html = urlopen('https://novel.naver.com/webnovel/finish?genre=all&order=Read&page={}'.format(page_num))
            self.soup = BeautifulSoup(html.read(), 'html.parser')
            self.naver_basic_info()
            
            #마지막 페이지 확인
            for islastPage in self.soup.find('div', {'class':'default_paging'}).children:
                allpage.append(islastPage.text)
            nowpagestr  = str(page_num)+'현재 페이지'
            if nowpagestr not in allpage:
                break
            page_num += 1
            sleep(0.4)
            
    #이미지url, 제목, 작가, 연재상태, 연재화수
    def naver_basic_info(self):
        thumbnail_all = self.soup.select('div#content > div > ul.card_list > li.item')

        #제목
        for i in thumbnail_all:
            self.photourl.append(i.select_one('img').attrs['src'])
            self.title.append(i.find('a').attrs['title']) #제목
            self.status.append(i.select_one('a > div > span.bullet_comp_ex').get_text()) #연재상태
            self.writer.append(i.select_one('span.author').get_text()) #연재상태
            self.status_num.append(i.select_one('span.count > span').get_text()) #연재화
            self.pagelink.append('https://novel.naver.com'+i.select_one('a.link').attrs['href'])
    #상세페이지 크롤링        
    
a= scraper()
a.naver()
```
